### モジュールのインポートとGoogleドライブのマウント

```python
# 画像処理モジュールのインポート
from PIL import Image

# 数値計算モジュールのインポート
import numpy as np

# Googleドライブをマウント
from google.colab import drive
drive.mount('/content/drive')
```

### グレースケール画像の読み込みと表示

```python
# グレースケール画像の読み込み
img_gray = Image.open('drive/MyDrive/python_image_recognition/data/coffee.jpg')

# グレースケール画像を表示
display(img_gray)
```

###グレースケール画像の配列の大きさと画素値の確認

```python
# グレースケール画像の配列の形を表示
print('グレースケール画像を保持する配列の形: {}'.
        format(np.array(img_gray).shape))

# グレースケール画像の原点の画素値を表示
print('グレースケール画像の(0, 0)における画素値: {}'.
        format(img_gray.getpixel((0, 0))))
```

### カラー画像の読み込みと表示

###カラー画像の配列の大きさと画素値の確認

```python
# カラー画像の配列の形を表示
print('カラー画像を保持する配列の形: {}'.
        format(np.array(img_color).shape))

# カラー画像の原点の画素値を表示
print('カラー画像の(0, 0)における画素値: {}'.
        format(img_color.getpixel((0, 0))))
```
